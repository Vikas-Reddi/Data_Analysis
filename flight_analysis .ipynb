```python
#import liblary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
file_path = "Airline Dataset Updated - v2.csv"
data = pd.read_csv(file_path)
```

```python
sns.set(style="whitegrid")
```

```python
gender_stats = data['Gender'].value_counts()
print("Gender-wise Passenger Distribution:\n", gender_stats)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_stats.index, y=gender_stats.values, palette='pastel')
plt.title("Gender-wise Passenger Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# 3. Airport Statistics (Top 10 Airports)
airport_stats = data['Airport Name'].value_counts().head(10)
print("\nTop 10 Airports by Traffic:\n", airport_stats)

plt.figure(figsize=(8, 5))
airport_stats.plot(kind='bar', color='lightgreen')
plt.title("Top 10 Airports by Traffic")
plt.xlabel("Airport Name")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Total Flights Managed by Each Pilot
pilot_stats = data['Pilot Name'].value_counts().head(10)
print("\nTop 10 Pilots with Most Flights:\n", pilot_stats)

plt.figure(figsize=(8, 5))
pilot_stats.plot(kind='bar', color='skyblue')
plt.title("Top 10 Pilots by Number of Flights")
plt.xlabel("Pilot Name")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Continental Flight Statistics
continent_stats = data['Airport Continent'].value_counts()
print("\nFlight Distribution by Continent:\n", continent_stats)

plt.figure(figsize=(6, 4))
continent_stats.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Continental Flight Distribution")
plt.ylabel("")  # Hide the y-label for aesthetics
plt.show()
```

```python
# 5. Pie Chart of Flight Status
flight_status_stats = data['Flight Status'].value_counts()
print("\nFlight Status Distribution:\n", flight_status_stats)

plt.figure(figsize=(6, 4))
flight_status_stats.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('muted'))
plt.title("Flight Status Distribution")
plt.ylabel("")  # Hide the y-label for aesthetics
plt.show()
```
